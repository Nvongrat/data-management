Hypothesis: Songs with 'dance' in the title are mostly songs with high danceability (.7 or higher)

Why .7?: Because the danceability score's maximum is 1.0, .7 seems like a reasonable minimum qualifier

Source: Spotify (see search_url)

```python
import requests
import pandas as pd
import base64
import json
import urllib
```

```python
search_url = 'https://api.spotify.com/v1/search?q={}&type={}&limit=50'.format(urllib.parse.quote('dance'), 'track')
```

```python
CID = pd.read_csv('Spotify_Keys.txt')['Client_ID'].iloc[0]
CS = pd.read_csv('Spotify_Keys.txt')['Client_Secret'].iloc[0]
```

```python
client_cred = base64.b64encode(str(CID + ":" + CS).encode("ascii"))
```

```python
headers = {"Authorization": "Basic {}".format(client_cred.decode("ascii"))}
```

Below is the endpoint for this analysis. This was chosen for easy access to Spotify data.

```python
payload = {'grant_type' : 'client_credentials'}
url = 'https://accounts.spotify.com/api/token'
```

```python
session_key_response = requests.post(url = url, data = payload, headers = headers)
```

```python
header_key = session_key_response.json()
```

```python
key = header_key['access_token']
```

```python
session_headers = {"Authorization": "Bearer {}".format(key)}
```

Data compilation starts now

```python
search_response = requests.get(url = search_url, headers = session_headers)
```

```python
data = search_response.json()
```

```python
search_df = pd.DataFrame(data['tracks']['items'])
```

Below is a the compilation of all data needed to complete the analysis

```python
search_df
```

```python
audio_features = 'https://api.spotify.com/v1/audio-features?ids={}'.format(','.join(list(search_df['id'])))
```

```python
audio_features
```

The data points below are the features of each item that will be focused on.

```python
features_response = requests.get(url = audio_features, headers = session_headers)
```

```python
features = features_response.json()
```

```python
features
```

```python
features_df = pd.DataFrame(features['audio_features'])
```

```python
joined_df = pd.merge(search_df, features_df, how = 'inner', on = 'id')
```

```python
joined_df
```

The dataframes below are used to single out the names and 'danceability' of the selected tracks.

Names are shown for proof that all data points qualify

```python
results = pd.DataFrame(joined_df['danceability'])
```

```python
result_names = pd.DataFrame(joined_df['name'])
```

```python
result_names
```

```python
results
```

```python
result_names['danceability'] = list(results['danceability'])
```

#Reflection  

This analysis is limited by the scale of Spotify's archives and methods of holding data. The ratings used for the analysis are able to change over time, and there is no way to get a maximum or minimum rating for any of the ratings that are compiled. There is also a language barrier, as all of the data points have English titles. Checking international songs with the same parameters seems like the next step.
